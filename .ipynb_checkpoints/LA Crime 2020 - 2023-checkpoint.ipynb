```python
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv('../data_course/Crime_Data_from_2020_to_Present.csv')

# Suppress the SettingWithCopyWarning
pd.options.mode.chained_assignment = None
```

# Understand and Clean the Data

In this section, we perform initial data exploration and cleaning steps. This includes examining the data schema, checking for null values, removing irrelevant columns, converting data types, and filtering out invalid data points.

## Key observations:
 - The dataset contains 27 columns with information about crime incidents.
 - Several columns have high null values and are removed for analysis.
 - Date columns are converted to datetime format for easier manipulation.
 - Invalid victim ages (negative or 0) are filtered out.


```python
crime.head()
```

```python
crime.columns
```

```python
crime.shape
```

```python
crime.dtypes
```

```python
#Check for Null values
crime.isna().sum()
```

```python
#Remove high null and irrelevant columns
df = crime.drop(labels = ['modus_operandi', 'weapon_description',\
                          'weapon_code', 'crime_code_2',\
                          'crime_code_3', 'crime_code_4',\
                          'cross_street'], axis = 1)
```

```python
df
```

```python
df.sort_values(by = 'date_reported')
```

```python
#Convert objects to datetimes where relevant and update datatypes
df['date_reported'] = pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
df['date_occurred'] = pd.to_datetime(df['date_occurred'], format = '%Y-%m-%d')
```

```python
df.info()
```

```python
df.describe(include = ['float64', 'int64'])
```

```python
# Extract the 'victim_age' column into a new DataFrame
df['victim_age'].value_counts()
```

```python
# Filter out invalid negative ages and age 0
df_age = df[df['victim_age'] > 0]
```

```python
df_age['victim_age'].value_counts()
```

```python
#Adding year and month columns
df['year'] = df['date_occurred'].dt.year
df['month'] = df['date_occurred'].dt.month
```

```python
df.sort_values(by = 'date_reported')
```

## Feature Understanding

### In this section, we explore and visualize various features of the crime data to gain insights into patterns and trends.

### Key findings:
 - The count of crime incidents varies by month, with higher counts generally observed in the middle months of the year.
 - There is an increasing trend in the number of crime incidents from 2020 to 2022.
 - The distribution of victim ages shows a peak in the 20-40 age range.
 - The top 20 most frequent crimes are identified, with theft-related crimes being the most prevalent.


```python
# Plot the count of crime incidents by month
df['month'].value_counts().sort_index().plot()
```

```python
# Plot the count of crime incidents by year as a bar chart
df['year'].value_counts().sort_index().plot(kind = 'bar')
```

```python
# Remove rows where month is 12 and year is 2023 (incomplete data)
df = df[~((df['month'] == 12) & (df['year'] == 2023))]
```

```python
# Sort the DataFrame by the 'date_occurred' column
df.sort_values(by = 'date_occurred')
```

```python
# Plot the count of crime incidents by year and month
crime_over_time = df.groupby(['year', 'month']).size().plot()
crime_over_time
```

```python
# Plot a histogram of victim ages
plt.hist(df_age['victim_age'], edgecolor = 'black', bins = 15)
```

```python
df['date_occurred_no_time'] = df['date_occurred'].dt.date
```

```python
#Calculate average daily count of crimes
crime_count_per_day = df.groupby(['date_occurred_no_time', 'crime_description']).size()
```

```python
#Calculate top 20 daily crimes
top_20_crimes = crime_count_per_day.groupby('crime_description').mean()\
.reset_index(name = 'average_daily_count').\
sort_values(by = 'average_daily_count', ascending = False).head(20)
```

```python
top_20_sorted = top_20_crimes.sort_values(by = 'average_daily_count', ascending = True)
```

```python
#Plot the top 20 cri'0omes
plt.barh(top_20_sorted['crime_description'], top_20_sorted['average_daily_count'])
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize = (12, 8))

# Plot 1, line chart
plt.subplot(2, 2, 1)
crime_plot = df.groupby(['year', 'month']).size().plot(color = 'green', linewidth = 2)
crime_plot.set_xlabel('Year - Month', fontsize = 10)
crime_plot.set_title('Crime Over Time')
# Plot 2, histogram of victim ages
plt.subplot(2, 2, 2)
plt.hist(df_age['victim_age'], edgecolor = 'black', bins = 15)
plt.xlabel('Age', fontsize = 10)
plt.tight_layout()
plt.title('Distribution of Victim Age')
#Plot 3, top 20 crimes
plt.figure()
plt.barh(top_20_sorted['crime_description'], top_20_sorted['average_daily_count'], color = 'orange')
plt.title('Top 20 Crimes')
```

## Time Taken to Report Crime


### In this section, we analyze the time difference between when a crime occurred and when it was reported.

### Key observations:
 - A large number of crimes are reported on the same day they occur (time difference of 0 days).
 - Some crime types have longer average reporting times than others.


```python
# Show the difference in time for crime reported vs occurred
df['Time to report'] = (pd.to_datetime(df['date_reported']) - pd.to_datetime(df['date_occurred_no_time'])).dt.days
```

```python
df
```

```python
df['Time to report'].value_counts().head(10)
```

```python
#Group by crime description and reporting times
df.groupby(['crime_description'])['Time to report'].mean().sort_values(ascending = False).head(10)
```

### Identify the top 3 crimes with the highest average victim age

```python
df_age.groupby('crime_description')['victim_age'].mean().reset_index().sort_values(by = 'victim_age', ascending = False).head(3)
```

## Areas With the Highest Crime Rates at Night

In this section, we identify the areas with the highest crime rates during nighttime hours (8 PM to 6 AM).

### Key finding:
 The top five areas with the highest nighttime crime rates are:

- Central (55,315 crimes)
- 77th Street (51,137 crimes)
- Pacific (47,889 crimes)
- Southwest (45,718 crimes)
- Hollywood (42,763 crimes)


The areas with the lowest nighttime crime rates are Foothill (27,267 crimes) and Hollenbeck (30,549 crimes). There is a significant variation in nighttime crime rates across different areas of Los Angeles.

```python
crimes_at_night = df[(df['date_occurred'].dt.hour) >= 20 | (df['date_occurred'].dt.hour <= 6)]
```

```python
crimes_at_night['area_name'].value_counts().reset_index()
```

```python
crimes_at_night_counts = crimes_at_night['area_name'].value_counts().reset_index()
crimes_at_night_counts.columns = ['area_name', 'count']

plt.figure(figsize=(10, 8))
plt.barh(crimes_at_night_counts['area_name'], crimes_at_night_counts['count'])
plt.xlabel('Number of Crimes')
plt.ylabel('Area Name')
plt.title('Number of Crimes at Night by Area')
plt.tight_layout()
plt.show()
```

## Crime by the Hour

In this section, we analyze the distribution of crimes across different hours of the day.

### Key observations:
 - There is a notable peak in crime incidents around noon (12 PM).
 - Identity theft is a significant contributor to the crime count at this hour.

```python
hourly_crime = df['date_occurred'].dt.hour.value_counts().sort_index().reset_index()
hourly_crime.columns = ['hour', 'crime_count']
```

```python
hourly_crime
```

```python
hourly_crime.columns
```

```python
#Crimes including identity theft
plt.bar(hourly_crime['hour'], hourly_crime['crime_count'])
```

```python
df[df['date_occurred'].dt.hour == 12]['crime_description'].value_counts()
```

```python
#Filter out identity theft to focus on active crime reporting
no_id_theft = df[~(df['crime_description'] == 'THEFT OF IDENTITY')]
```

```python
new_hourly_count = no_id_theft['date_occurred'].dt.hour.value_counts().sort_index().reset_index()
```

```python
new_hourly_count.columns = ['hour', 'crime_count']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(new_hourly_count['hour'], new_hourly_count['crime_count'])
plt.xlabel('Hour')
plt.ylabel('Crime Count (Excluding Identity Theft)')
plt.title('Crime by the Hour (Excluding Identity Theft)')
plt.xticks(new_hourly_count['hour'])
plt.show()
```
